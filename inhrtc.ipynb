```python
balance=1000
class grandparent():
    def deposit(self):
        self.amount=int(input("enter amount to deposit : "))
        self.balance=self.amount+1000
        print("new balnce = ", self.balance)


class parent(grandparent):
    def withdraw(self):
        self.amount=int(input("enter withdrawing amount : "))
        if self.amount<balance:
            print("your new balance is = ", self.balance - self.amount )
        else:
           
            print("insufficient balance!!!")
            
        


class child(parent):
    def view(self):
        print("your current balance is = ", self.balance )


c=child()

c.deposit()
c.withdraw()



```

```python
#multiple inheritwnce



class A():
    def add(self):
        print('addition',5+5)

class B():
    def sub(self):
        print('subtraction',10-5)

class C(A,B):
    def mul(self):
        print('multiplication',5*5)

a=A()
b=B()
c=C()


c.add()
c.sub()
c.mul()
```
