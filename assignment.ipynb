```python
class BankAccount():

    def __init__(self):
        self.balance=1000

    def deposite(self):
        amount = int(input('enter your amount to deposite : '))
        totalamount =amount+ self.balance
        self.balance = totalamount

        print('Your updated balance is : ', totalamount)     
    
    
    def withdraw(self):
        amount = int(input('enter your amount to withdraw : '))

        if amount <= self.balance:

           self.balance -= amount
           print('Your balance afte withdrawing is : ',  self.balance)     
        else:
            print('Insuffecent balance!!')


    
f=BankAccount()
        
while True:

        print('''
            1. Deposite
            2. Withdraw
            3. Exit    ''')

        choice = int(input('enter your choice : '))

        if choice == 1:
            f.deposite()

        elif choice == 2:
            f.withdraw()

        
        elif choice == 3:
            print('Exit....')
            break

        else:
            print('Invalid choice!!')
            choice = int(input('enter your choice : '))
```

```python
#polymorphism 
class dog:
    def sound(self):
        print('bark')

class cat():
    def sound(self):
        print('meou')

class bird():
    def sound(self):
        print("chirp")

class goat():
    def sound(self):
        print("mehh...")


a=dog()
b=cat()
c=bird()
d=goat()

a.sound()
b.sound()
c.sound()
d.sound()
```
